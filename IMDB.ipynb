```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
r=requests.get("https://www.imdb.com/chart/top/")
r
```

```python
soup=BeautifulSoup(r.content,"lxml")
soup.prettify()
```

```python
filmler=[]
```

```python
films= soup.find_all("tbody",{"class":"lister-list"})

for film in films:
    film=film.find_all("td",{"class":"titleColumn"})
    for fim in film:
        isim=fim.find("a").text
        filmler.append(isim)
        
        
        
    
        
        
```
